```python
import pandas as pd
import requests
import numpy as np
from matplotlib import pyplot as plt
```

```python
file= "Resources/project1_data.csv"
df = pd.read_csv(file)
df.head()
```

```python
new_df= df[df["IndicatorCode"]=="SDGAIRBODA"]
new_df.head()
```

```python
del new_df['IndicatorCode']
del new_df['Indicator']
del new_df['ValueType']
del new_df['ParentLocationCode']
del new_df['Location type']
del new_df['Period type']
del new_df['IsLatestYear']
del new_df['Dim1 type']
del new_df['Dim3 type']
del new_df['FactValueUoM']
del new_df['FactValueNumericLowPrefix']
del new_df['FactValueNumericHighPrefix']
del new_df['FactValueTranslationID']
del new_df['FactComments']
del new_df['Language']
del new_df['DateModified']
del new_df['Dim3']
del new_df['Dim3ValueCode']
del new_df['DataSourceDimValueCode']
del new_df['DataSource']
del new_df['FactValueNumericPrefix']
del new_df['Dim2 type']
del new_df['Dim2ValueCode']
del new_df['SpatialDimValueCode']
del new_df['Dim1ValueCode']
new_df.head()
```

```python
new_df=new_df.rename(columns= {'ParentLocation':'Continent', 'FactValueNumeric': 'Average Death Rate', 'Period': 'Year', 'Dim1':'Sex', 'FactValueNumericLow': 'Regional Low Deaths', 'FactValueNumericHigh': 'Regional High Deaths',
                      'Dim2': 'Cause of Death'})
new_df.head()
```

```python
#Quartiles
death_rate= both_sexes['Average Death Rate']
quartiles = death_rate.quantile([.25,.5,.75])
lowerq = quartiles[0.25]
upperq = quartiles[0.75]
iqr = upperq-lowerq

print(f"The lower quartile of average death rate is: {lowerq}")
print(f"The upper quartile of average death rate is: {upperq}")
print(f"The interquartile range of average death rate is: {iqr}")
print(f"The the median of average death rate is: {quartiles[0.5]} ")

lower_bound = lowerq - (1.5*iqr)
upper_bound = upperq + (1.5*iqr)
print(f"Values below {lower_bound} could be outliers.")
print(f"Values above {upper_bound} could be outliers.")
```

```python
# Find lowest death by country for TBLC

tblc_deaths = new_df.loc[new_df['Cause of Death'] == 'Trachea, bronchus, lung cancers']
total_tblc_deaths = tblc_deaths.sort_values('Average Death Rate', ascending= True)
total_tblc_deaths.head()
```

```python
# Find lowest death by country for TBLC for both sexes
tblc_bthsex_country = total_tblc_deaths.loc[total_tblc_deaths['Sex'] == 'Both sexes']
tblc_bthsex_country = tblc_bthsex_country.sort_values('Average Death Rate', ascending= True)
tblc_bthsex_country = tblc_bthsex_country.reset_index(drop= True)
low_bthsex_tblc_country = tblc_bthsex_country.loc[0:4, :]
low_bthsex_tblc_country
```

```python
# Countries with Lowest TBLC deaths for both sexes
x_axis = np.arange(len(low_bthsex_tblc_country))
# tick_locations = [value for value in x_axis]



x_tblc_bthsex = low_bthsex_tblc_country['Location']
y_tblc_bthsex = low_bthsex_tblc_country['Average Death Rate']
plt.barh(x_tblc_bthsex, y_tblc_bthsex, color = 'orange')
plt.xlabel('Average Death Rate per 100,000')
plt.ylabel('Countries')
plt.title('Countries with the Lowest Trachea, Bronchus, and Lung Cancer Death Rate for Both Sexes')
plt.xticks(min(x_axis), max(x_axis)+.5)


plt.tight_layout()

# Set x and y limits
plt.ylim(-0.75, len(x_tblc_bthsex))
plt.xlim(0.0,.5)

plt.show()
```

```python
# Find lowest death by country for TBLC for females
tblc_fem_country = total_tblc_deaths.loc[total_tblc_deaths['Sex'] == 'Female']
tblc_fem_country = tblc_fem_country.sort_values('Average Death Rate', ascending= True)
tblc_fem_country = tblc_fem_country.reset_index(drop= True)
low_fem_tblc_country = tblc_fem_country.loc[0:4, :]
low_fem_tblc_country
```
